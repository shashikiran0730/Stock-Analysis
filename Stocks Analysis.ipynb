## Top Companies Based on Roc

```python
df.sort_values(by='ROCE % ',ascending=False,)[['Name','ROCE % ']].head()
```

```python
df.sort_values(by='ROCE % ',ascending=False,)[['Name','ROCE % ']].head(10).plot(kind='bar',x='Name',title='Name vs top 10 company ROC',figsize=(10,5))
```

## Top  Companies based on Profit Growth

```python
df.sort_values(by='Qtr Profit Var % ',ascending=False,)[['Name','Qtr Profit Var % ']].head(10)
```

```python
df.sort_values(by='Qtr Profit Var % ',ascending=False,)[['Name','Qtr Profit Var % ']].head(10).plot(x='Name',title='Name Vs Top 10 Qtr Profit Var %',kind='barh',figsize=(15,5))
plt.xlabel('Qtr Profit Var')
plt.show()
```

## Top Companies based on Sales Growth

```python
df.sort_values(by='Qtr Sales Var %',ascending=False,)[['Name','Qtr Sales Var %']].head(5).plot(x='Name',title='Name Vs Top 5 Qtr Qtr Sales',kind='line',figsize=(15,5))
plt.ylabel('Qtr Sales Var')
plt.show()
```

```python
df.sort_values(by='EV / EBITDA',ascending=False,)[['Name','EV / EBITDA']].head(5).plot(x='Name',title='Name Vs Top 5 Qtr Qtr Sales',kind='line',figsize=(15,5))
plt.ylabel('Qtr Sales Var')
plt.show()
```

## Top companies Based on Market CAP
